```python
import pandas as pd
import matplotlib
import seaborn
from tqdm import tqdm
from datetime import datetime

def convert_to_timestamp(date_string):
    months_array = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
    formats = ["%d/%m/%Y", "%d %b, %Y", "Q%q %Y", "%d %b %Y", "%d %b, %Y"]
    for fmt in formats:
        for month in months_array:
            fmt = fmt.replace("%b", month)
            try:
                dt = datetime.strptime(date_string, fmt)
                return dt.timestamp()
            except ValueError:
                pass
    return None  # Return None if date format is not recognized
```

```python
other_games_df = pd.read_csv("./unnecessary/games_clean_info_stream2.csv")
other_games_df = other_games_df.drop("About", axis="columns").drop("Supported Languages", axis="columns")

progress = tqdm(range(len(other_games_df)))
for index, row in other_games_df.iterrows():
    if row["Genres"] == "[]":
        other_games_df = other_games_df.drop(index=index)
    progress.update()
progress.close()
# games_df
other_games_df.to_csv("other_games.csv")
```
